```python
prob_exang = df['exang'].mean() * 100
print(f"Probability of exercise-induced angina: {prob_exang:.2f}%")
```

```python
prob_high_chol = df[df['chol'] > 240]['target'].mean() * 100
print(f"Probability of heart disease given cholesterol > 240: {prob_high_chol:.2f}%")
```

```python
def chol_function(row):
    if row['chol'] > 240:
        return 'High'
    elif 200 <= row['chol'] <= 240:
        return 'Borderline'
    else:
        return 'Normal'


df['chol_level'] = df.apply(chol_function, axis=1)
df.head()
```

```python
df['bp_level'] = df['trestbps'].apply(lambda x: 'Hypertension' if x > 140 else ('Pre-Hypertension' if 140 >= x >= 120 else 'Normal'))
df.head()
```

```python
df['thalch_result'] = df.apply(lambda row: 1 if row['thalach'] < 187 else 0, axis=1)
df.head()
```

```python
df['hr_reserve'] = 220 - df['age'] - df['thalach']
df['hr_reserve'] = df['hr_reserve'].apply(lambda x: 'Low' if x < 70 else ('Borderline' if 70 <= x <= 85 else 'Normal'))
df.head()

```

```python
df['chol_age'] = df['chol'] / df['age']
df['chol_age'] = df['chol_age'].apply(lambda x: 'High' if x > 5 else ('Borderline' if 3 <= x <= 5 else 'Normal'))
df.head()
```

```python

df['high_risk'] = df.apply(
    lambda row: 1 if (
        row['slope'] == 2 and
        row['chol'] > 240 and
        row['exang'] == 1 and
        (row['cp'] == 0 or row['cp'] == 3)
    ) else 0,
    axis=1
)
df.head()


```
